```python
import pylab as pl
import pandas as pd
import matplotlib.pyplot as plt
pl.style.use('ggplot')

%pylab inline
```

```python
# The data is yelp data which contain 1197 resturants in Pittsburgh combining with Census data merged by zipcodes.
# each row represent a resturant
# 'Competitorcont' coloum represent the number of resturants in the same zipcode area and and as same categories. 
df = pd.read_csv('newdata.csv')
df = df.drop('Unnamed: 0', 1)
df = df[df['review_count'] > 100]
df
```

```python
df.dtypes
```

```python
#plot 

pl.figure(figsize(15,10))

groups = df.groupby('categories')

fig, ax = plt.subplots()
ax.margins(0.05) # adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.review_count, group.stars, marker='o', linestyle='', ms=20, label=name, alpha=0.8)

ax.legend()
pl.xlabel('Review Count', fontsize = 20)
pl.ylabel('Stars', fontsize = 20)
pl.title('Yelp Stars of Resturants by Review Count and Categories', fontsize = 20)
pl.xlim(80,500)

plt.show()

```
