```python
from __future__ import print_function
__author__ = 'Tianqi Jiang'
from numpy import random as rd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
distributions={'normal':{},'poisson':{},'binomial':{},'chisquare':{},'gamma':{},'exponential':{}}
```

```python
def generate_size():
    return rd.random_integers(10,2000,1)[0]
```

```python
mymean = 100 
for i in xrange(100):
    distributions['normal'][i] = rd.normal(mymean,1,generate_size())
    distributions['poisson'][i] = rd.poisson(mymean,generate_size())
    distributions['binomial'][i] = rd.binomial(2*mymean,0.5,generate_size())
    distributions['chisquare'][i] = rd.chisquare(mymean,generate_size())
    distributions['gamma'][i] = rd.gamma(mymean,1,generate_size())
    distributions['exponential'][i] = rd.exponential(mymean, generate_size())
```

```python
def plot_mean(name):
    plt.figure(figsize=(10,5))
    for i in distributions[name]: 
        plt.plot(len(distributions[name][i]),distributions[name][i].mean(),'o')
    plt.xlabel('sample size')
    plt.ylabel('sample mean')
    plt.plot([20,2000],[100,100],'r')
    plt.title('mean of %s distribution'%(name), color='black')
    plt.show()
```

```python
for name in distributions.keys():
    plot_mean(name)
```

# Just as expected, when the sample size become larger, the sample mean trend to close to 100.

```python
allmeans=[]
for i in distributions:
    for j in distributions[i]:
        allmeans.append(distributions[i][j].mean())
```

```python
plt.hist(allmeans,bins=10)
plt.xlabel('sample mean')
plt.ylabel('N')
plt.title('histogram of all means')
```
